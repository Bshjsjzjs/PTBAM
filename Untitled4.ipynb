```python
import numpy as np

def matrix_inverse_3x3(matrix):
    # Convert the input to a NumPy array
    mat = np.array(matrix, dtype=float)

    # Check if the matrix is 3x3
    if mat.shape != (3, 3):
        return "Error: Input matrix must be 3x3."

    # Calculate the determinant
    det = np.linalg.det(mat)

    # Check for singularity
    if abs(det) < 1e-9:  # Use a small threshold for floating point comparison
        return "The matrix is singular, so its inverse does not exist."
    else:
        # Calculate the inverse
        inverse_mat = np.linalg.inv(mat)
        return inverse_mat

# Example usage:
# Non-singular matrix
A = [[1, 2, 3],
     [0, 1, 4],
     [5, 6, 0]]
print("Matrix A:")
print(np.array(A))
print("\nInverse of A:")
print(matrix_inverse_3x3(A))

print("\n" + "-" * 30 + "\n")

# Singular matrix (determinant is 0)
B = [[1, 2, 3],
     [4, 5, 6],
     [7, 8, 9]]
print("Matrix B:")
print(np.array(B))
print("\nInverse of B:")
print(matrix_inverse_3x3(B))
```
